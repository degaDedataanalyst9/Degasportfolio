```python
#transform objects into dummies
df = pd.get_dummies(data = df, drop_first = True)        # pd is used here because it is the library used for data manipulation and analysis
df.head()
```

#MULTILINEAR REGRESSION PREPERATION


```python
#ISOLATE X AND Y
y = df.salary / 1000
x = df.drop(columns = ["salary","yrs.service"])
x.head()
```

```python
# add a constant
x = sm.add_constant(x)
x.head()
```

```python
# training and test  set
from sklearn.model_selection import train_test_split
X_train, X_test ,Y_train,Y_test = train_test_split(x,y,
                                                   test_size = 0.2,
                                                   random_state = 1502)
X_train.head()
```

#Multi-Linear Regression


```python
#multilinear regression
model = sm.OLS(Y_train,X_train).fit()
print(model.summary())
```

# Accuracy assesment

```python
#predict
predictions = model.predict(X_test)
predictions[:4]
```

```python
# Accuracy assesment
from sklearn.metrics import mean_absolute_error,mean_squared_error
print("MAE:", round(mean_absolute_error(Y_test,predictions),1))
print("RSME:", round(np.sqrt(mean_squared_error(Y_test,predictions)),1))
```
