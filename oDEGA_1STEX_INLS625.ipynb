```python
# An excel csv file with rows(from the powerpoint) transposed as coloumns was created.
# The data frame was assigned to the csv file created.
df = pd.read_csv("Dega_Class1_INLS625_exe.csv")
df.head(30)
```

```python
# using fillna to replace "NaN" values to smothen calculations
df = df.fillna(0)
df
```

```python
# .describe gives a snapshot of all mean,min,max for all the coloums in data frame
df.describe()
```

```python
# Clculating Mean,mode,median,max & min using libraries for row_1
df.COL1.mean()
```

```python

data_without0 = df[df != 0]
data_without0
print("row_1 mean:",df.COL1.mean())
print("row_1 mode:",data_without0.COL1.mode())
print("row_1 median:",df.COL1.median())
print("row_1 max:",df.COL1.max())
print("row_1 min:",data_without0.COL1.min())
```

```python
from collections import Counter
# Mathemetical way for mean @ row_1
# assigning a data frame called df_1 to the coloumn 1
df_1 = data_without0.iloc[:,0]
mean_1 = sum(df_1)/len(df_1)
mean_1
# Mathemetical way for mode @ row_1
df1_list = list(df_1)
df1_list
frequency1 = Counter(df1_list)
frequency1
# for loops
for num,count in frequency1.items():
  if count == max(frequency1.values()):
    print("num in list:",num,"frequency of the number:",count,"row_1 mode:",num)

```
